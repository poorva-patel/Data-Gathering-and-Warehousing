All we know at the start of our investigation is the <b>murder</b> occurred on <b>Jan.15, 2018</b> and that it took place in <b>SQL City<b>.

```r
#Write a query in SQL to view the crime scene report 
#for the murder occurred on Jan.15, 2018 and that it took place in SQL City.
#Select ALL records FROM the table WHERE...
csr <- dbSendQuery(db, "SELECT *
                        FROM crime_scene_report
                        WHERE
                            date = '20180115'
                        AND city = 'SQL City'
                        AND type = 'murder'
                        ")
dbFetch(csr)
```

Another clue! It looks like we have two witnesses! Now who are they and how do we find them?

```r
#Write two queries in SQL to find each witness in the person table. 
#For the first query as first_witness, select all the records from person where the address_street_name 
#is Northwestern Drive and order by adress_number
#Fetch the first_witness
first_witness <- dbSendQuery(db, "SELECT * 
                        FROM person
                        WHERE address_street_name = 'Northwestern Dr' 
                        ORDER BY address_number
                        ")
dbFetch(first_witness)
```

We have identified our first witness as Morty Schapiro. Now lets find the second!

```r
#For the third query as second_witness,
#Select all the records from person where the address_street_name is Frnaklin Ave and the name includes Annabel. 
#Fetch the second_witness
second_witness <- dbSendQuery(db, "SELECT * 
                        FROM person
                        WHERE address_street_name = 'Franklin Ave' 
                        AND name LIKE '%Annabel%'
                        ")
dbFetch(second_witness)
```

```r
#Create a query to select for the following clues: 

suspects <- dbSendQuery(db, "SELECT *
                             FROM get_fit_now_member
                        ") 
dbFetch (suspects) #Found 3 members whose membership starts with 48Z 

suspect <- dbSendQuery(db, "SELECT *
                             FROM drivers_license
                            WHERE plate_number LIKE '%H42W%'
                        ")
dbFetch (suspect) #Found 2 male owners with the following license characters

time <- dbSendQuery(db, "SELECT *
                             FROM get_fit_now_check_in
                            ORDER BY check_in_time, check_in_date
                        ")
dbFetch (time) #Found 2 people

query_combined <- " SELECT p.id, p.name
                    FROM person p
                    LEFT JOIN get_fit_now_member m ON (p.id = m.person_id)
                    LEFT JOIN get_fit_now_check_in c ON (m.id = c.membership_id)
                    LEFT JOIN drivers_license d ON (p.license_id = d.id)
                    WHERE c.check_in_date = '20180109'
                    AND c.membership_id LIKE '48Z%'
                    AND d.plate_number LIKE '%H42W%';
                  "
result <- dbSendQuery(db, query_combined) #showing the combined results
data <- dbFetch(result)
print(data)
```

Three different people match this description! Now which one attended the SQL Symphony Concert three times? 

```r
#Let's find the killer!
#checking people who attended
murderer <- dbSendQuery(db, " SElECT *
          FROM facebook_event_checkin
          WHERE event_name LIKE '%Concert%'
          ")

dbFetch(murderer)

concert <- " SELECT p.id, p.name, d.height, d.hair_color, d.gender, d.car_make, d.car_model
                                FROM person p
                                LEFT JOIN drivers_license d ON (p.license_id = d.id)
                                WHERE d.height IN (65, 66, 67) AND d.hair_color = 'red'
                                AND d.car_make = 'Tesla' AND d.car_model = 'Model S'
                                AND p.id IN (
                                    SELECT f.person_id
                                    FROM facebook_event_checkin f
                                    WHERE f.event_name = 'SQL Symphony Concert'
                                    AND f.date LIKE '201712%'
                                    GROUP BY f.person_id
                                    HAVING COUNT(f.person_id) = 3);
                                "
attendees <- dbSendQuery(db, concert) #combine previous query with new info
attendees_data <- dbFetch(attendees)
print(attendees_data)
```
